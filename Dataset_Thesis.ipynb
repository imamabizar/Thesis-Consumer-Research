```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(pwd + "/Dataset_Thesis.xlsx")
dataset
```

```python
dataset_copied = dataset.copy()
dataset_copied
```

```python
id_vars = list(dataset_copied.columns) [ : 9]
value_vars = list(dataset_copied.columns) [9 : ]
```

```python
dataset_melted = dataset_copied.melt(id_vars=id_vars, value_vars=value_vars, var_name="question_id + no", value_name="Answer")
dataset_melted
```

```python
question_import = pd.read_excel(pwd + "/Dataset_Thesis.xlsx", sheet_name="Question_Edited")
question_import
```

```python
questions = question_import.copy()
questions.drop(columns=["question_id", "no", "questions"], inplace=True)
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=questions, how="left", left_on="question_id + no", right_on="question_id + no")
print("Original Data", len(dataset_melted))
print("Merged Data", len(dataset_merged))
dataset_merged
```

```python
respondents = dataset_merged[dataset_merged["Answer"].notna()]
respondents = respondents.groupby("question")["Timestamp"].nunique().reset_index()
respondents.rename(columns={"Timestamp":"Respondents"}, inplace=True)
respondents
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondents, how="left", left_on="question", right_on="question")
print("Original Data", len(dataset_merged))
print("Merged Data", len(dataset_merged_two))
dataset_merged_two
```

```python
same_answer = dataset_merged # [dataset_merged["Answer"].notna()]
same_answer = same_answer.groupby(["question_id + no", "Answer"])["Timestamp"].nunique().reset_index()
same_answer.rename(columns={"Timestamp":"Same Answer"}, inplace=True)
same_answer
```

```python
dataset_merged_three = pd.merge(left=dataset_merged_two, right=same_answer, how="left", left_on=["question_id + no", "Answer"], right_on=["question_id + no", "Answer"])
dataset_merged_three["Same Answer"].fillna(0, inplace=True)
print("Original Data", len(dataset_merged_two))
print("Merged Data",len(dataset_merged_three))
dataset_merged_three
```

```python
dataset_merged_three.to_excel(pwd + "/Final_OutputThesis.xlsx", index=False)
```
